```python
## High order functions #
# - functions that can take other functions as arguments or return functions as results

def add(x, y):
    return x + y    
def subtract(x, y):
    return x - y
def operate(func, x, y):
    return func(x, y)
result1 = operate(add, 5, 3)
result2 = operate(subtract, 5, 3)
print(f"Addition: {result1}")
print(f"Subtraction: {result2}")

```

```python
## map()- applies a function to all items in an iterable and returns a map object 
numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x**2, numbers)
print(list(squared))
```

```python
def square(x):
    return x**2

numbers = [1, 9, 3, 4, 5]
squared = map(square, numbers)
print(list(squared))
```

```python

## filter() - filters items in an iterable based on a function that returns True or False
# keep only when cond satisfied
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))

```

```python
def is_even(x):
    return x % 2 == 0

numbers = [1, 2, 3, 4, 5]
even_numbers = filter(is_even, numbers)
print(list(even_numbers))

```

```python
##reduce - applies a rolling computation to sequential pairs of values in a list
#  reducing it to a single value
from functools import reduce    
numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

```python
from functools import reduce
def add(x,y):
    return x + y
numbers = [1, 2, 3, 4, 5]   
sum = reduce(add, numbers)
print(sum)
```

```python
#decorators - functions that modify the behavior of other functions without changing their code
def decorator(func):
    def wrapper():
        print("Before the function is called.")
        func()
        print("After the function is called.")
    return wrapper  
```

```python
@decorator
def say_hello():
    print("Hello!")
say_hello()


print("\n")

@decorator
def say_goodbye():
    print("Goodbye!")   
say_goodbye()
```

```python
#validators - functions that check if the input data is valid and meets certain criteria
def validate_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    elif age > 120:
        raise ValueError("Age cannot be greater than 120.")
    else:
        print(f"Valid age: {age}")
try:
    validate_age(25)
    validate_age(-5)
    validate_age(1500)
except ValueError as e:
    print(e)

## validator using decorators

def validate_age(func):
    def wrapper(age):
        if age < 0:
            raise ValueError("Age cannot be negative.")
        elif age > 120:
            raise ValueError("Age cannot be greater than 120.")
        else:
            return func(age)
    return wrapper


@validate_age
def print_age(age):
    print(f"Valid age: {age}")
try:
    print_age(25)
    print_age(-5)
    print_age(1500) 
except ValueError as e:
    print(e)
    
```

```python
## Generators - functions that return an iterator and can be used to iterate over a sequence of values

def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
counter = count_up_to(5)
for number in counter:
    print(number) 



```
