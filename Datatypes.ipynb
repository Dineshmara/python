## Datatypes In Python

```python
## integer (int) data type
a = 10
print(a)  # Output: 10
print(type(a))  # Output: <class 'int'>
## float data type
b = 3.14
print(b)  # Output: 3.14
print(type(b))  # Output: <class 'float'>
## string (str) data type
c = "Hello, World!"
print(c)  # Output: Hello, World!
print(type(c))  # Output: <class 'str'>     
## boolean (bool) data type
d = True
print(d)  # Output: True
print(type(d))  # Output: <class 'bool'>
```

```python
## list - ordered,mutable,heterogeneous collection of items

my_list = [1, "hello", 3.14, True]
print(my_list)  # Output: [1, 'hello', 3.14, True]
print(type(my_list))  # Output: <class 'list'>
print(my_list[0])  # Output: 1
print(my_list[1])  # Output: hello  
updateed_list = [1, "hello", 3.14, True]
updateed_list[1] = "world"
print(updateed_list)  # Output: [1, 'world', 3.14, True]        


## tuple - ordered,immutable,heterogeneous collection of items
my_tuple = (1, "hello", 3.14, True)
print(my_tuple)  # Output: (1, 'hello', 3.14, True)
print(type(my_tuple))  # Output: <class 'tuple'>    
print(my_tuple[0])      
print(my_tuple[1])  # Output: hello
modifying_tuple = (1, "hello", 3.14, True)
# modifying_tuple[1] = "world"  # This will raise a TypeError because tuples are immutable


## set - unordered,mutable,heterogeneous collection of unique items and
## does not allow duplicate values
my_set = {1, "hello", 3.14, True}   
print(my_set)  
print(type(my_set))
my_set.add("world")
print(my_set)  # Output: {1, 'hello', 3.14, True, 'world'}
my_set.remove(3.14)
print(my_set)  # Output: {1, 'hello', True, 'world'}



## dictionary - unordered,mutable,heterogeneous collection of key-value pairs
my_dict = {"name": "Alice", "age": 30, "is_student": True}
print(my_dict)  # Output: {'name': 'Alice', 'age': 30, 'is_student': True}
print(type(my_dict))  # Output: <class 'dict'>
print(my_dict["name"])  # Output: Alice
print(my_dict["age"])  # Output: 30
my_dict["age"] = 31 
print(my_dict)  # Output: {'name': 'Alice', 'age': 31, 'is_student': True}



```
