## LOOPS in Python

```python
## LOOPS - A loop is a programming construct that allows you to repeat a block of code multiple times. In Python, there are two main types of loops: for loops and while loops.

numbers = [1, 2, 3, 4, 5]
 
# For loop example
for number in numbers:
    print(number)   

# While loop example
count = 0
while count < 5:
    print(count)
    count += 1  

A= [x for x in numbers]
print(A)

## patterns
for i in range(5):
    for j in range(i + 1):
        print("*", end="")
    print() 

for i in range(5):
    for j in range(5 - i):
        print("*", end="")
    print()

for i in range(5):
    for j in range(5):
        if j < 5 - i - 1:
            print(" ", end="")
        else:
            print("*", end="")
    print()


for i in range(5):
    if i == 0 or i == 4:
        print("*" * 5)
    else:
        print("*" + " " * 3 + "*")


## while loop pattern
i = 0
while i < 5:
    j = 0
    while j < i + 1:
        print("*", end="")
        j += 1
    print()
    i += 1 

i = 0
while i < 5:
    j = 0
    while j < 5 - i:
        print("*", end="")
        j += 1
    print()
    i += 1
```
