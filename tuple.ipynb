## Tuples

```python
## Tuples in Python - A tuple is an ordered, immutable collection of items. #
# It is defined using parentheses `()` and can contain elements of different data types.#
#  Once a tuple is created, its contents cannot be modified.

Tuple1 = (1, 2, 3, 4, 5)
print(Tuple1)

tuple_methods = [method for method in dir(Tuple1) if not method.startswith('_')]
print("Tuple methods:", tuple_methods)

print(Tuple1[0])  # Accessing the first element of the tuple
print(Tuple1[1:4])  # Slicing the tuple to get a subset

print(len(Tuple1))  # Finding the length of the tuple
print(max(Tuple1))  # Finding the maximum value in the tuple
print(min(Tuple1))  # Finding the minimum value in the tuple

print(Tuple1 + (6, 7, 8))  # Concatenating two tuples
print(Tuple1 * 2)  # Repeating the tuple

print(Tuple1.count(3))  # Counting the occurrences of an element in the tuple
print(Tuple1.index(4))  # Finding the index of an element in the tuple


## lambda 

B = list(map(lambda x: x**2, Tuple1))  # List of lambda functions
print(B)
```
