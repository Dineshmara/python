setdefault() , fromkeys(), pyramid pattern in while loop


```python
## setdefault()
# The setdefault() method in Python is used to set a default value for a key in a dictionary if the key doesn't already exist.
# It returns the value of the key if it exists, otherwise it sets and returns the default value.

my_dict = {"a": 1, "b": 2}
default_value = my_dict.setdefault("c", 3)
print(default_value)  
print(my_dict)   

default_value_existing = my_dict.setdefault("a", 10)
print(default_value_existing) 
print(my_dict)

## fromkeys()
# The fromkeys() method in Python is used to create a new dictionary with keys from an iterable and values set to a default value.
keys = ["x", "y", "z"]
default_value = 0
new_dict = dict.fromkeys(keys, default_value)
print(new_dict)

## Pyramid pattern using while loop
rows = 5
i = 1
while i <= rows:
    print(" " * (rows - i) + "*" * (2 * i - 1))
    i += 1

```

5 codes on map(), filter(), reduce(), decorators, lambda functions

```python


def square(x):
    return x**2 
numbers = [1, 9, 3, 4, 5]
squared = map(square, numbers)
print(list(squared))

def multiply(x, y):
    return x * y
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
result = map(multiply, numbers1, numbers2)
print(list(result))



def is_positive(x):
    return x > 0
numbers = [-2, -1, 0, 1, 2]
positive_numbers = filter(is_positive, numbers)
print(list(positive_numbers))


nums = [5,3,4,6,6,]
print(filter(lambda x: x**2, nums))  # Output: 25
print(list(filter(lambda x: x**2, nums)))  # Output: [5]

def is_even(x):
    return x % 2 == 0
numbers = [1, 2, 3, 4, 5]

even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))



from functools import reduce
def add(x,y):
    return x + y
numbers = [1, 2, 3, 4, 5]
sum = reduce(add, numbers)
print(sum)

def multiply(x,y):
    return x * y    
numbers = [1, 2, 3, 4, 5]
product = reduce(multiply, numbers) 
print(product)



def decorator(func):
    def wrapper():
        print("Before the function is called.")
        func()
        print("After the function is called.")
    return wrapper

@decorator
def say_hello():
    print("Hello!")
say_hello()
square_lambda = lambda x: x**2
print(square_lambda(5))

@decorator
def say_goodbye():
    print("Goodbye!")
say_goodbye()

@decorator
def say_welcome():
    print("Welcome!")
say_welcome()



```

Generators using for loop

```python
##Generators are a type of iterable in Python that allow you to iterate over a sequence of values without storing them all in memory at once. They are created using generator functions, which use the yield keyword to produce a value and pause the function's execution until the next value is requested.
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
counter = count_up_to(5)
for number in counter:
    print(number)

print("\n")

def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b
fib_sequence = fibonacci(10)
for num in fib_sequence:
    print(num)

print("\n")

def even_numbers(n):
    for i in range(n):
        if i % 2 == 0:
            yield i
evens = even_numbers(20)
for even in evens:
    print(even) 
```
