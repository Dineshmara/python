```python
## setdefault()
# The setdefault() method in Python is used to set a default value for a key in a dictionary if the key doesn't already exist.
# It returns the value of the key if it exists, otherwise it sets and returns the default value.

my_dict = {"a": 1, "b": 2}
default_value = my_dict.setdefault("c", 3)
print(default_value)  
print(my_dict)   

default_value_existing = my_dict.setdefault("a", 10)
print(default_value_existing) 
print(my_dict)

## fromkeys()
# The fromkeys() method in Python is used to create a new dictionary with keys from an iterable and values set to a default value.
keys = ["x", "y", "z"]
default_value = 0
new_dict = dict.fromkeys(keys, default_value)
print(new_dict)

## Pyramid pattern using while loop
rows = 5
i = 1
while i <= rows:
    print(" " * (rows - i) + "*" * (2 * i - 1))
    i += 1

```

5 codes on map(), filter(), reduce(), decorators, lambda functions

```python


def square(x):
    return x**2 
numbers = [1, 9, 3, 4, 5]
squared = map(square, numbers)
print(list(squared))

def multiply(x, y):
    return x * y
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
result = map(multiply, numbers1, numbers2)
print(list(result))



def is_positive(x):
    return x > 0
numbers = [-2, -1, 0, 1, 2]
positive_numbers = filter(is_positive, numbers)
print(list(positive_numbers))


nums = [5,3,4,6,6,]
print(filter(lambda x: x**2, nums))  # Output: 25
print(list(filter(lambda x: x**2, nums)))  # Output: [5]

def is_even(x):
    return x % 2 == 0
numbers = [1, 2, 3, 4, 5]

even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))



from functools import reduce
def add(x,y):
    return x + y
numbers = [1, 2, 3, 4, 5]
sum = reduce(add, numbers)
print(sum)

def multiply(x,y):
    return x * y    
numbers = [1, 2, 3, 4, 5]
product = reduce(multiply, numbers) 
print(product)



def decorator(func):
    def wrapper():
        print("Before the function is called.")
        func()
        print("After the function is called.")
    return wrapper

@decorator
def say_hello():
    print("Hello!")
say_hello()
square_lambda = lambda x: x**2
print(square_lambda(5))

@decorator
def say_goodbye():
    print("Goodbye!")
say_goodbye()

@decorator
def say_welcome():
    print("Welcome!")
say_welcome()



```

```python
## Exception handling


''' 1. SyntaxError: 
    2. NameError:
    3. TypeError: 
    5. IndexError: 
    6. KeyError: 
    8. ImportError: '''



try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")

try:
    result= -1
except ValueError:
    print("Value must be non-negative!")

try:
    result = int("abc")
except ValueError:
    print("Invalid input! Please enter a number.")      

try:
    print("Hello")
except SyntaxError:
    print("SyntaxError: Missing closing parenthesis.")

try:
    result = "5" + 10
except TypeError:
    print("TypeError: Cannot concatenate str and int.")
try:   
    my_list = [1, 2, 3]
    print(my_list[5])
except IndexError:
    print("IndexError: List index out of range.")   

try:
    my_dict = {"a": 1, "b": 2}
    print(my_dict["c"])
except KeyError:
    print("KeyError: 'c' not found in dictionary.")         



## Exception handling in functions
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero!"
print(divide(10, 2))
print(divide(10, 0))


def get_item(lst, index):
    try:
        return lst[index]
    except IndexError:
        return "Index out of range!"
my_list = [1, 2, 3]
print(get_item(my_list, 1))
print(get_item(my_list, 5))

def read_file(filename):
    try:
        with open(filename, 'r') as file:
            return file.read()
    except FileNotFoundError:
        return "File not found!"
print(read_file("existing_file.txt"))
print(read_file("non_existent_file.txt"))   

```

Generators using for loop

```python
##Generators are a type of iterable in Python that allow you to iterate over a sequence of values without storing them all in memory at once. They are created using generator functions, which use the yield keyword to produce a value and pause the function's execution until the next value is requested.
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
counter = count_up_to(5)
for number in counter:
    print(number)

print("\n")

def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b
fib_sequence = fibonacci(10)
for num in fib_sequence:
    print(num)

print("\n")

def even_numbers(n):
    for i in range(n):
        if i % 2 == 0:
            yield i
evens = even_numbers(20)
for even in evens:
    print(even) 
```

```python
#Encapsulation is a fundamental principle of object-oriented programming that restricts access to certain components of an object and bundles the data and methods that operate on that data within a single unit, or class. It helps to protect the internal state of an object from unintended interference and misuse, while still allowing controlled access through public methods. This promotes modularity, maintainability, and security in software design.

## eaxmples of encapsulation,inheritance, polymorphism, abstraction, and method overriding are all fundamental concepts in object-oriented programming (OOP) that help to structure and organize code in a way that promotes reusability, maintainability, and scalability. Here are examples of each concept:

class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."    
    
class Dog(Animal):
    def speak(self):
        return f"{self.name} says: Woof!"   
class Cat(Animal):
    def speak(self):
        return f"{self.name} says: Meow!"
dog = Dog("Buddy")
cat = Cat("Whiskers")
print(dog.speak())  
print(cat.speak())

##Inheritance is a fundamental principle of object-oriented programming that allows a new class from parent
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def start_engine(self):
        return f"{self.make} {self.model} engine started."  
class Car(Vehicle):
    def start_engine(self):
        return f"{self.make} {self.model} car engine started."
class Motorcycle(Vehicle):
    def start_engine(self):
        return f"{self.make} {self.model} motorcycle engine started."
    
car = Car("Toyota", "Camry")
motorcycle = Motorcycle("Harley-Davidson", "Street 750")    
print(car.start_engine())
print(motorcycle.start_engine())

## Abstraction is a fundamental principle of object-oriented programming that focuses on hiding the complex implementation details of a system and exposing only the essential features and functionalities to the user. It allows developers to create simplified models of real-world entities, making it easier to manage and understand complex systems. Abstraction helps to reduce complexity, improve code maintainability, and enhance the overall design of software applications.

class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement this method.")
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        import math
        return math.pi * self.radius ** 2
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
circle = Circle(5)
rectangle = Rectangle(4, 6)
print(f"Area of the circle: {circle.area()}")
print(f"Area of the rectangle: {rectangle.area()}")

##Method overriding is a feature of object-oriented programming that allows a subclass to provide a specific implementation of a method that is already defined in its superclass. When a method in a subclass has the same name, return type, and parameters as a method in its superclass, the method in the subclass overrides the method in the superclass. This allows the subclass to modify or extend the behavior of the inherited method while still maintaining the same interface.
class Parent:
    def greet(self):
        return "Hello from the Parent class!"
class Child(Parent):
    def greet(self):
        return "Hello from the Child class!"
parent = Parent()
child = Child()
print(parent.greet())
print(child.greet())

    
```
