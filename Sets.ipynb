## SETS In Python

```python
##sets - A set is an unordered collection of unique elements. #
# It is defined using curly braces `{}` or the `set()` constructor. #
# Sets are mutable, meaning you can add or remove elements from them.

# Set methods in python
""" 1. add(): Adds an element to the set.
    2. remove(): Removes an element from the set. Raises a KeyError if the element is not found.
    3. discard(): Removes an element from the set if it is present. Does not raise an error if the element is not found.
    4. pop(): Removes and returns an arbitrary element from the set. Raises a KeyError if the set is empty.
    5. clear(): Removes all elements from the set.
    _______________________________________________________________________________________________
    
    6. union(): Returns a new set that is the union of two sets.
    7. intersection(): Returns a new set that is the intersection of two sets.
    8. difference(): Returns a new set that is the difference of two sets.
    9. symmetric_difference(): Returns a new set that is the symmetric difference of two sets.
    10. issubset(): Returns True if the set is a subset of another set.
    11. issuperset(): Returns True if the set is a superset of another set.
    12. isdisjoint(): Returns True if the set has no elements in common with another set."""

set1={1, 2, 3, 4, 5}
set2={4, 5, 6, 7, 8,5}
set1.add(8)
print(set1)
set1.remove(2)
print(set1)
set1.discard(10)  # This will not raise an error since 10 is not in the set
print(set1)
set1.pop()  # This will remove and return an arbitrary element from the set
print(set1)
print("\n")


set1={1, 2, 3, 4, 5}
set2={4, 5, 6, 7, 8,5}

print(set1.union(set2))  # This will return a new set that is the union of set1 and set2
print(set1.intersection(set2))  # This will return a new set that is the intersection of set1 and set2
print(set1.difference(set2))  # This will return a new set that is the difference of set1 and set2
print(set1.symmetric_difference(set2))  # This will return a new set that is the symmetric difference of set1 and set2
print(set1.issubset(set2))  # This will return True if set1 is a subset of set2
print(set1.issuperset(set2))  # This will return True if set1 is a superset of set2


set1={1, 2, 3, 4, 5}
set2={ 6, 7, 8}
print(set1.isdisjoint(set2))  # This will return True if set1 has no elements in common with set2           
```
