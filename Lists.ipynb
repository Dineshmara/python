## Lists in Python


````python
## python Lists = ordered, changeable, allows duplicate members
mylist = ["apple", "banana", "cherry", "apple"]
print(mylist)
print(mylist[1])  

mylist[1] = "blueberry"  
print(mylist)

mylist.append("orange")  
print(mylist)

mylist.insert(1, "grape") 
print(mylist)

mylist.remove("apple")  # Removing an item from the list
print(mylist)

mylist.pop(2)  # Removing the last item from the list
print(mylist)

mylist.sort()  # Sorting the list
print(mylist)

mylist.reverse()  # Reversing the list
print(mylist)

print(len(mylist))  # Getting the length of the list
print(mylist.count("apple"))  # Counting the occurrences of "apple" in the list
print(mylist.index("apple"))  # Finding the index of "apple" in the list
mylist.extend(["kiwi", "melon","apple"])  # Extending the list with another list
print(mylist)
 
for i in mylist:  # Looping through the list
    if i.isalpha():  # Checking if the item is a string
        mylist.remove(i)  
print(mylist)


````

## Uses of NumPy

NumPy is a powerful library for numerical computing in Python. Here are some common uses:


**Example:**

```python
import numpy as np

# Create a 1D array
arr = np.array([1, 2, 3, 4, 5])
print('Array:', arr)

# Element-wise operations
print('Squared:', arr ** 2)

# Statistical operations
print('Mean:', np.mean(arr))
print('Standard Deviation:', np.std(arr))

# 2D array and matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print('Matrix Product:', np.dot(A, B))
```
