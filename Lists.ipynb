## Lists in Python


```python
## python Lists = ordered, changeable, allows duplicate members
mylist = ["apple", "banana", "cherry", "apple"]
print(mylist)
print(mylist[1])  # Accessing the second item in the list

mylist[1] = "blueberry"  # Changing the second item in the list
print(mylist)

mylist.append("orange")  # Adding a new item to the end of the list
print(mylist)

mylist.insert(1, "grape")  # Inserting an item at a specific position
print(mylist)

mylist.remove("apple")  # Removing an item from the list
print(mylist)

mylist.pop(2)  # Removing the last item from the list
print(mylist)

mylist.sort()  # Sorting the list
print(mylist)

mylist.reverse()  # Reversing the list
print(mylist)

print(len(mylist))  # Getting the length of the list
print(mylist.count("apple"))  # Counting the occurrences of "apple" in the list
print(mylist.index("apple"))  # Finding the index of "apple" in the list
mylist.extend(["kiwi", "melon","apple"])  # Extending the list with another list
print(mylist)
 
for i in mylist:  # Looping through the list
    if i.isalpha():  # Checking if the item is a string
        mylist.remove(i)  
print(mylist)
```
