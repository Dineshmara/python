## Uses of Dictionary

Python dictionaries are unordered collections of keyâ€“value pairs. Common uses include:



**Examples:**

```python
# Basic dictionary
person = {"name": "Alice", "age": 30}
print(person["name"])  # fast lookup

# Iteration
for k, v in person.items():
    print(k, v)

# Counting with dict.get
words = ["apple", "banana", "apple"]
counts = {}
for w in words:
    counts[w] = counts.get(w, 0) + 1
print(counts)  # {'apple': 2, 'banana': 1}

# Using collections.defaultdict for counts
from collections import defaultdict
counts2 = defaultdict(int)
for w in words:
    counts2[w] += 1
print(dict(counts2))

# Nested dictionaries (JSON-like)
data = {
    "user1": {"score": 10},
    "user2": {"score": 15}
}
print(data["user2"]["score"])  # 15

# Dictionary comprehension
squares = {x: x*x for x in range(6)}
print(squares)

# Safe lookup with get and default
print(person.get("address", "Unknown"))
```


Dictionary in Python

````python
dictionary = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}


dictionary.update({"city": "USA"})  # Adding a new key-value pair to the dictionary
print(dictionary)
## enumerate() function is used to get the index and value of each item in the dictionary
for i,j in enumerate(dictionary.items(),1):
    print(i,j)

print(dictionary["name"])  # Accessing the value associated with the key "name"
print(dictionary.get("age"))  # Accessing the value associated with the key "age" using the get method
print(dictionary.keys())  # Getting all the keys in the dictionary
print(dictionary.values())  # Getting all the values in the dictionary
print(dictionary.items())  # Getting all the key-value pairs in the dictionary

print(list(dictionary.items()))
````
